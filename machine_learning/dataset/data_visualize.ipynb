```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python

def draw_angled_rec(loc, ax, color, linewidth, label):
    y0, x0, width, height, angle = loc

    # 计算长方形的四个角的坐标
    angle_rad = math.radians(angle)
    cos_theta = math.cos(angle_rad)
    sin_theta = math.sin(angle_rad)
    vertices = np.array([
        [x0 - width / 2 * cos_theta - height / 2 * sin_theta, y0 + height / 2 * cos_theta - width / 2 * sin_theta],
        [x0 + width / 2 * cos_theta - height / 2 * sin_theta, y0 + height / 2 * cos_theta + width / 2 * sin_theta],
        [x0 + width / 2 * cos_theta + height / 2 * sin_theta, y0 - height / 2 * cos_theta + width / 2 * sin_theta],
        [x0 - width / 2 * cos_theta + height / 2 * sin_theta, y0 - height / 2 * cos_theta - width / 2 * sin_theta]
    ])

    # 创建Polygon对象，并将其添加到图形中
    polygon = Polygon(vertices, closed=True, edgecolor=color, linewidth=linewidth, fill=None)
    ax.text(x0, y0, str(label), color='black', fontsize=8, ha='center', va='center')
    ax.add_patch(polygon)


def map_to_img(data_input):
    default_scale = 530 / 0.34

    data_input[0] = data_input[0] * default_scale
    data_input[1] = data_input[1] * default_scale + 320
    data_input[2:4] *= default_scale
    data_input[4] -= 1.57
    data_input[4] *= -1
    return data_input
```

```python
objects_num = 4  # from 4-15
label_or_input = 'label'  # or "label"
dataset = np.loadtxt('./%s/num_%d.txt' % (label_or_input, objects_num))
show_example_id = 0
sample0 = dataset[show_example_id]
fig, ax = plt.subplots()
objects_num = len(sample0) // 5

for i in range(objects_num):
    loc = sample0[i * 5:i * 5 + 5]
    loc = map_to_img(loc)
    draw_angled_rec(loc, ax, 'black', 3, i)

ax.set_aspect('equal')
ax.autoscale()
ax.set_ylim(0, 480)
ax.set_xlim(0, 640)
plt.show()
```
